```python
df.reset_index(drop=True, inplace=True)
df.head()
```

```python
from surprise import Dataset, NormalPredictor, Reader
from surprise.model_selection import cross_validate


# A reader is still needed but only the rating_scale param is requiered.
reader = Reader(rating_scale=(1, 10))

# The columns must correspond to user id, item id and ratings (in that order).
data = Dataset.load_from_df(df[["AS_1", "AS_2", "rating"]], reader)

# We can now use this dataset as we please, e.g. calling cross_validate
cross_validate(NormalPredictor(), data, cv=2)
```

```python
from surprise.model_selection import KFold
from surprise import SVD, KNNBasic, accuracy

# define a cross-validation iterator
kf = KFold(n_splits=3)

algo = SVD()

for trainset, testset in kf.split(data):

    # train and test algorithm.
    algo.fit(trainset)
    predictions = algo.test(testset)

    # Compute and print Root Mean Squared Error
    accuracy.rmse(predictions, verbose=True)
```

```python
predictions
```

```python
pred_1 = algo.predict(str('AS_3'), str('AS_8'))
pred_2 = algo.predict(str('AS_8'), str('AS_3'))
pred_3 = algo.predict(str('AS_0'), str('AS_7'))
pred_4 = algo.predict(str('AS_7'), str('AS_0'))
```

```python
print(f"Prediction(uid='{pred_1.uid}', iid='{pred_1.iid}', est={round(pred_1.est, 2)})")
print(f"Prediction(uid='{pred_2.uid}', iid='{pred_2.iid}', est={round(pred_2.est, 2)})")
print(f"Prediction(uid='{pred_3.uid}', iid='{pred_3.iid}', est={round(pred_3.est, 2)})")
print(f"Prediction(uid='{pred_4.uid}', iid='{pred_4.iid}', est={round(pred_4.est, 2)})")
```

```python
as_list = [f"AS_{i}" for i in range(11)]
```

```python
full_ratings = pd.DataFrame(columns=[i for i in ['AS'] + as_list])
```

```python
full_ratings
```

```python
def get_value(as_1, as_2):
    try:
        return df[(df['AS_1'] == as_1) & (df['AS_2'] == as_2)]['rating'].values[0]
    except:
        if as_1 == as_2:
            return None
        return round(algo.predict(str(as_1), str(as_2)).est)
```

```python
for curr_as in as_list:
    to_add = [curr_as] + [get_value(curr_as, other_as) for other_as in as_list]
    full_ratings = pd.concat([full_ratings, pd.DataFrame([to_add], columns=full_ratings.columns)])

full_ratings.reset_index(drop=True, inplace=True)
full_ratings.fillna(value=np.nan, inplace=True)
```

```python
full_ratings
```

```python
full_ratings.to_excel('full_svd.xlsx')
```

```python
full_ratings
```

```python
full_mean = full_ratings.drop(columns=['AS']).stack().mean()
```

```python
full_ratings_first = full_ratings.copy()
```

```python
for curr_as in as_list:
    full_ratings_first.loc[full_ratings_first[curr_as] < full_mean, curr_as] = 0
    full_ratings_first.loc[full_ratings_first[curr_as] >= full_mean, curr_as] = 1
    full_ratings_first[curr_as].fillna(value=1, inplace=True)
    full_ratings_first[curr_as] = full_ratings_first[curr_as].astype(np.int8)
```

```python
full_ratings_first
```

```python
full_ratings_first.to_excel('full_svd_allow_1.xlsx')
```

```python
full_ratings_second = full_ratings.copy()
full_ratings_second['means'] = full_ratings.drop(columns=['AS']).mean(axis=1)
```

```python
full_ratings_second
```

```python
for curr_as in as_list:
    full_ratings_second.loc[full_ratings_second[curr_as] < full_ratings_second['means'], curr_as] = 0
    full_ratings_second.loc[full_ratings_second[curr_as] >= full_ratings_second['means'], curr_as] = 1
    full_ratings_second[curr_as].fillna(value=1, inplace=True)
    full_ratings_second[curr_as] = full_ratings_second[curr_as].astype(np.int8)

full_ratings_second.drop(columns=['means'], inplace=True)
```
